```python
# importing libraries

# import scipy

# from sklearn.preprocessing import MinMaxScaler
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Step 2
# Load the dataset
df = pd.read_csv('diabetes.csv')
```

```python
# Step 6
# Check the outliers 
# Box plot
fig, axs = plt.subplots(9,1, figsize=(7,17))
i = 0
for col in df.columns:
    axs[i].boxplot(df[col], vert=False)
    axs[i].set_ylabel(col)
    i+=1
plt.show()



```

```python
# Step 7 
# Drop the outliers

clean_data = df.copy()  # Assuming df is the DataFrame containing the data

for column in df.columns:
    if column in ['SkinThickness','Outcome']: 
        continue
    q1, q3 = np.percentile(clean_data[column], [25, 75])
    iqr = q3 - q1
    factor = 1.5
    if column=='BloodPressure':
        factor = 0.75
    lower_bound = q1 - (factor * iqr)
    upper_bound = q3 + (factor * iqr)
    clean_data = clean_data[(clean_data[column] >= lower_bound) & (clean_data[column] <= upper_bound)]



```

```python
#correlation

sns.heatmap(df.corr(),annot =True, fmt= '.2f')
plt.show()
```
